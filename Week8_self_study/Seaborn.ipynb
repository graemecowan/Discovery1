# Seaborn: common chart types

### Imports

To use Seaborn effectively, we need to import it: it is often imported as `sns`. Here we will also import matplotlib, pandas and numpy since Seaborn is built on top of matplotlib and we require pandas and numpy to handle our example data.



````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

#### Palmer Penguins


This dataset contains information about penguins from three different species that were observed on the Palmer Archipelago, Antarctica. The data includes information about the penguins' species, island, bill length and depth, flipper length, body mass, and sex.

The data originally appeared in:

Gorman KB, Williams TD, Fraser WR (2014). Ecological sexual dimorphism and environmental variability within a community of Antarctic penguins (genus Pygoscelis). PLoS ONE 9(3):e90081. https://doi.org/10.1371/journal.pone.0090081

<sub>Photo: Paul Carroll via unslash: https://unsplash.com/photos/Y-nyDv3TWm0</sub>


````python
import pandas as pd

# Load the Palmer Penguins dataset
penguins = pd.read_csv('palmer_penguins.csv')
penguins.head()
````

### Histograms

Histograms are useful for visualizing the distribution of a single variable, such as bill length in millimeters in the penguins dataset.

````python
sns.histplot(data=penguins, x="bill_length_mm", kde=False)
plt.savefig("histogram.pdf")
````

# Bar plots

Bar plots are good for visualising values of a categorical and numerical variable together, such as the body mass of each penguin species.

````python
sns.barplot(data=penguins, x="species", y="body_mass_g")
plt.savefig("barplot.pdf")
````

Bar plots with hue can be used to show additional information about each group, such as the sex of each penguin species.

````python
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex")
plt.savefig("barplot_hue.pdf")
````

### Line plot

A line plot is typically used to show how a single variable changes over time or across some other continuous dimension. The line represents the trend or pattern in the data, with each point on the line indicating the value of the variable at a particular time or point in the dimension being plotted. Line plots are particularly useful for showing trends, patterns, or fluctuations in data over time or space.

````python
ax = sns.lineplot(data=penguins, x="year", y="body_mass_g")

# Set the x-ticks to the integer values of 'year'
ax.set_xticks(penguins['year'].unique())

# Save the plot as a PDF file
plt.savefig("lineplot.pdf")
````

### Scatter plots 

Scatter plots are similar to line plots, but can be used to visualize the relationship between two variables.

````python
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
plt.savefig("scatterplot.pdf")
````

### Swarm plot

Swarm plots can be used to show the distribution of data points within each group, such as the distribution of bill depths for each penguin species.

````python
sns.swarmplot(data=penguins, x="species", y="bill_depth_mm")
plt.savefig("swarmplot.pdf")
````

### Box plot

Box plots are useful for visualizing the distribution of data within each group and comparing groups, such as the distribution of bill lengths for each penguin species.

````python
sns.boxplot(data=penguins, x="species", y="bill_length_mm")
plt.savefig("boxplot.pdf")
````

### Violin plot
Violin plots are similar to box plots, but also show the density of data within each group.

````python
sns.violinplot(data=penguins, x="species", y="bill_length_mm")
plt.savefig("violinplot.pdf")
````

### Heatmaps

Heatmaps are useful for visualizing the correlation between multiple variables. Here since the penguin dataset doesn't have data that would be appropriate to plot directly as a heatmap, a correlation table of variable in the dataset has been calculated and plotted. 


````python
correlation = penguins.corr()
sns.heatmap(data=correlation, annot=True, cmap="coolwarm")
plt.savefig("heatmap.pdf")
````

### Pair plots

Pair plots are a useful way to visualize the relationships between multiple variables, and can be used to compare groups by including hue.



````python
sns.pairplot(data=penguins, hue="species")
plt.savefig("pairplot.pdf")
````

# Saving files

Seaborn provides a simple way to save plots to a variety of file formats using the savefig() function from the Matplotlib library.

The savefig() function can save plots to several file formats, including:

* PNG: Portable Network Graphics format, which is a lossless image format that supports transparency and is ideal for images with large areas of solid color.
* JPG/JPEG: Joint Photographic Experts Group format, which is a lossy image format that uses compression to reduce file size. It is suitable for photographs and other complex images, but may not be suitable for images with sharp edges or text.
* PDF: Portable Document Format, which is a vector-based format that is ideal for plots that will be printed or included in documents or presentations.
* SVG: Scalable Vector Graphics format, which is a vector-based format that is ideal for plots that will be displayed on websites or in applications that support vector graphics.

To save a Seaborn plot using savefig(), simply call the function and pass in the desired file name and file format as arguments. For example, to save a plot as a PNG file named "my_plot.png", use the following code:

````python
plt.savefig("my_plot.png", format="png")
````

By default, savefig() will save the entire figure, including any whitespace and borders. To save only the actual plot, you can use the `bbox_inches` parameter, which specifies the portion of the figure to save. For example, to save only the plot area as a PNG file named "my_plot.png", use the following code:

````python
plt.savefig("my_plot.png", format="png", bbox_inches="tight")
````

Additionally, you can control the resolution of the saved file using the dpi parameter. By default, the resolution is set to 100 dots per inch, but you can adjust this value to increase or decrease the quality of the saved image. For example, to save a plot with a resolution of 300 dots per inch as a PDF file named "my_plot.pdf", use the following code:

````python
plt.savefig("my_plot.pdf", format="pdf", dpi=300)
````

# Formatting

Seaborn provides a variety of formatting options that allow you to customize the appearance of your visualizations. Some of the most common formatting options include:

### X and Y axis titles

You can add titles to the X and Y axes using the set_xlabel() and set_ylabel() methods, respectively. For example:

````python
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm");
plt.xlabel("Bill Length (mm)");
plt.ylabel("Bill Depth (mm)");

````

### Plot title

You can add a title to the entire plot using the title() function. For example:

````python

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm");
plt.title("Bill Length vs Bill Depth");
````

X tick rotation: You can rotate the X axis tick labels using the xticks() function. For example, to rotate the X tick labels by 45 degrees:

````python

sns.barplot(data=penguins, x="species", y="body_mass_g");
plt.xticks(rotation=45);
````

### Legend location

The plt.legend() function in Matplotlib is used to create a legend for a plot. The loc parameter specifies the location of the legend on the plot. Here are some options for loc:

* loc="best": Matplotlib will try to choose the best location for the legend based on the size of the plot and the location of the data. This is the default option if loc is not specified.
* loc="upper left": Places the legend in the upper left corner of the plot.
* loc="upper right": Places the legend in the upper right corner of the plot.
* loc="lower left": Places the legend in the lower left corner of the plot.
* loc="lower right": Places the legend in the lower right corner of the plot.
* loc="center": Places the legend in the center of the plot.
* loc="center left": Places the legend in the center of the left edge of the plot.
* loc="center right": Places the legend in the center of the right edge of the plot.
* loc="lower center": Places the legend in the center of the bottom edge of the plot.
* loc="upper center": Places the legend in the center of the top edge of the plot.

These options allow you to position the legend in various locations on the plot. For example, plt.legend(loc="lower right") would place the legend in the lower right corner of the plot. 

You can also use a combination of loc and bbox_to_anchor parameters to position the legend precisely, outside of the plot, or to control the spacing between the legend and the plot.

The bbox_to_anchor parameter of the plt.legend() function in Matplotlib allows you to specify an anchor point for the legend that is outside the plot area. The anchor point is specified as a tuple of two values, (x, y), where (0, 0) is the lower left corner of the plot and (1, 1) is the upper right corner of the plot.

e.g. ```plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))```
    

````python

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species");
plt.legend(loc="lower right");
````

# Figure size

You can adjust the size of the figure by changing the values passed to the figsize parameter of the figure() function. For example, ```plt.figure(figsize=(10, 8))``` would create a figure with a width of 10 inches and a height of 8 inches.

Note that you should call `plt.figure(figsize=(...))` **before** creating any plots in the notebook, as this sets the size of the entire figure. If you call this function after creating a plot, it will create a new empty figure with the specified size.


````python
# Set the figure size
plt.figure(figsize=(8, 6));

````
