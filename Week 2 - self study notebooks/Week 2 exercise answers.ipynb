```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#1. Open the dataset 'gapminder.csv'

gap_df=pd.read_csv('gapminder.csv')

display (gap_df.head())

#2. 

print(gap_df.year.unique())

#3. 

gap_df_2007 = gap_df.query("year == 2007")

#4. 

sns.scatterplot(data=gap_df_2007, x="gdpPercap", y="lifeExp", hue="continent")
plt.xlabel('GDP per capita ($US)');
plt.ylabel('Life Expectancy (years)');


```
